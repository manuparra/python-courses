<h2 style="colour: olive">Controlling the elements of our chart</h2>.

Matplotlib offers a great deal of freedom to configure options for the chart itself, such as titles, legend, axis limits, etc. A great strength of matplotlib is that it includes an engine to render Latex codes, which we can include in any part of our chart.

```python
# We import the necessary modules
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h3 style="colour: navy">Chart titles and axis labels</h3>.

The following functions allow you to specify titles for the chart and its x and y axes:

>- `title(title)` Title of the chart. Accepts a text string.
>- `xlabel(label)` Title of the x-axis. Accepts a text string.
>- `ylabel(label)` y-axis title. Accepts a text string.
>
> Text strings can incorporate latex codes (we will enclose them in 'dollar' symbols '$'. If our text string has accents or special characters, we weight a u in front of it (unicode).


```python
# Load the data from "data/iris.dat".
# This data comes from the Fisher dataset available at 
# http://archive.ics.uci.edu/ml/datasets/Iris
# The dataset has the following comma-separated columns
# 1. sepal length in cm 
# 2. sepal width in cm 
# 3. petal length in cm 
# 4. petal width in cm 
# 5. class: Iris Setosa - Iris Versicolour - Iris Virginica
# Plots a graph with columns 1 and 2 with different colours for each class
# Add chart titles and labels for the axes


labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colores = ["r", "b", "g"]
iris_data = np.genfromtxt("data/iris.dat", delimiter=",", dtype=None)
clases = iris_data["f4"]

for idx, label in enumerate(labels):
    s_lengths = iris_data["f0"]
    s_widths = iris_data["f1"]
    ids = clases == label
    plt.scatter(s_lengths[ids], s_widths[ids], marker="o", color=colores[idx])

plt.title("Fisher's Iris Dataset")
plt.xlabel("Sepal length (cm)")
plt.ylabel("Petal length (cm)")
```

```python
# Create a graph of the polynomial f(x) = 1/4 * (x+4)(x+1)(x-2) 
# between the values of x [-4, 4].
# Represent it with a green line
# Add title to graph and labels to axes
x = np.linspace(-4, 4, 25)
y = (1.0/4) * (x+4)*(x+1)*(x-2)
plt.plot(x,y, color = "g")
plt.title('$f(x)=\\frac{1}{4}(x+4)(x+1)(x-2)$')
plt.xlabel("X")
plt.ylabel("Y")
```

<h3 style="colour: navy">Leyenda</h3>

To add labels to the lines we represent with plot, we will specify the **label** argument when we call plot. When we call the **legend** function, it will display the specified line style and label.

To display the legend, we will use the **plt.legend(args, kwargs)** function. If we have specified the *label* argument these will be displayed in the legend, otherwise it is possible to specify that it will be displayed using an array of lines and an array of labels.

> ` `plt.legend(*args, **kwargs)`.
>- The arguments (if specified) will be a list of lines and a list of labels as follows: `plt.legend([line1, line2, ..., linen], [label1, label2, ..., labeln]`.
>- Some of the keyword arguments we can use are:
> - **loc**: Integer specifying the location of the legend. Possible codes are; ['best':0, 'upper right':1, 'upper left': 2, 'lower left':3, 'lower right':4, 'right':5, 'centre left':6, 'centre right':7, 'lower centre':8, 'upper center':9, 'centre':10]

> - **bbox_to_anchor**: Tuple with the coordinates (0-1) of the caption location. For example, to place the caption in the middle of the right axis we will use:
> - `loc='upper right', bbox_to_anchor=(0.5, 0.5)`.
> - **ncol** : Entry specifying the number of columns of the legend (1 by default)
    The number of columns that the legend has. Default is 1.
> - **fontsize** : Size of the legend font.
> - **frameon** : None or (True/False). Specifies whether to draw a box around the legend. 
> - **fancybox** : None or (True/False). Specifies whether the caption box has rounded edges. (only if frameon == True)
> - **shadow** : None or (True/False). Specifies whether the caption box will have a shadow.
> - **framealpha** : Float specifying the transparency of the caption box.
> - **title** : Text string with the caption title.

```python
# Create a representation of the following functions between [0, 100].
# y1 = x^2 -4x +10 --> dashed red line and blue points
# y2 = -x^2 +4x + 10 --> blue dashed line and red dots
# Add a title to the graph and a legend

x = np.linspace(0, 100, 20)
y1 = x**2 - 4*x + 10
y2 = -x**2 + 4*x + 10
func_label_1 = "$f(x)=x^2 -4x + 10$"
func_label_2 = "$f(x)=-x^2 +4x + 10$"
plt.plot(x, y1, c="r", mfc="b", marker="o", mec="k", label=func_label_1)
plt.plot(x, y2, c="b", mfc="r", marker="o", mec="k", label=func_label_2)
plt.title(u"Representación de dos polinómios")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(title=u"Polinómios", shadow=True, loc=3, fancybox=True)
```

```python
# In this example we are going to call the legend in a different way.
# We will store the lines in variables and use them when calling the legend.

line1, = plt.plot(x, y1, c="r", mfc="b", marker="o", mec="k")
line2, = plt.plot(x, y2, c="b", mfc="r", marker="o", mec="k")
plt.title(u"Representación de dos polinómios")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend((line1, line2), (func_label_1, func_label_2), title=u"Polinómios", 
           shadow=True, loc=3, fancybox=True)
```

```python
# Repeat the above chart but take the legend out of the chart area
# Use bbox_to_anchor property

line1, = plt.plot(x, y1, c="r", mfc="b", marker="o", mec="k")
line2, = plt.plot(x, y2, c="b", mfc="r", marker="o", mec="k")
plt.title(u"Representación de dos polinómios")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend((line1, line2), (func_label_1, func_label_2), title=u"Polinómios", 
           shadow=True, loc='lower left', bbox_to_anchor=(1, 0))
```

#### Tick control with the ticker module

The ticker module contains classes to control the location and format of ticks. To change the location and format of ticks we can use the functions of the Axe object:

- `axe.xaxis.set_major_locator(locator)`: Accepts a *matplotlib.ticker.locator* object for the main ticks.
- `axe.xaxis.set_minor_locator(locator)`: Accepts a `matplotlib.ticker.locator` object for secondary ticks.
- `axe.xaxis.set_major_formatter(formatter)`: Accepts a "fomatter" object for main ticks.
- `axe.xaxis.set_minor_formatter(formatter)`: Accepts a "formatter" object for secondary ticks.

To create locator objects, we can use the functions of the ticker module:
- `ticker.MultipleLocator(d)`: Creates ticks at the interval specified by d.
- `ticker.FixedLocator(positions)`: Creates ticks at the positions specified in the positions list.
- `ticker.FixedFormatter(namelist)`: creates labels according to a list of names
- `ticker.FuncFormatter(function)`: creates labels according to a function

```python
# Represents the normalised and denormalised cardinal sine functions between -20 and 20
# https://en.wikipedia.org/wiki/Sinc_function
# Numpy provides the normalised function np.sinc()
# For the denormalised function we will use sin(X)/X
# Uses on the X-axis ticks greater than 5 and less than 1
# Uses on the Y axis ticks greater than 0.2 and less than 0.1
import matplotlib.ticker as ticker
X = np.linspace(-20, 20, 1024)
Y1 = np.sinc(X)
Y2 = np.sin(X)/X

# To use these ticker module functions, we must retrieve the Axe object.
ax = plt.axes()
# We change the ticks of X
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.1))
plt.plot(X, Y1, c = 'r')
plt.plot(X, Y2, c = "b")
```

```python
# Create a simple bar chart with the values and labels from the following lists
values = (25, 50, 78, 30, 28)
labels = ('Vicente', 'Maria', 'Alejandro', 'Alberto', 'Jose')

pos_list = np.arange(len(labels))
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.FixedLocator((pos_list)))
ax.xaxis.set_major_formatter(ticker.FixedFormatter((labels)))

# Now with center
plt.bar(pos_list, values, color='r', align='center')
```

```python
# Plots the functions y = e^-10x and y = e^-5x between 0 and 1
# Uses a black line and a dashed black line
# Use ticker.FuncFormatter() to format the x-axis labels according to make_label()
def make_label(value, pos):
    return str(value * 100) + "%"

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(make_label))
X = np.linspace(0, 1, 100)
plt.plot(X, np.exp(-10 * X), c ='k')
plt.plot(X, np.exp(-5 * X), c= 'k', ls = '--')
```

```python
# Rotates the above labels by 30 degrees
# Use the get_xtickslabels() method of the Axe object to retrieve the labels.
# Use the set_rotation(degrees) method of each label object to rotate them
def make_label(value, pos):
    return str(value * 100) + "%"

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(make_label))
X = np.linspace(0, 1, 100)
plt.plot(X, np.exp(-10 * X), c ='k')
plt.plot(X, np.exp(-5 * X), c= 'k', ls = '--')

# Rotate labels 30 degrees
labels = ax.get_xticklabels()
for label in labels:
    label.set_rotation(30.)
```

```python
# The following dictionary contains the amounts of sugar (%) in various foods.
# Make a bar chart with title and axis labels.
# The y-axis labels should incorporate "%".
food_data = {"Ketchup":8.8, "Penaut Butter":9.2, 
             "Chocolate Bar":33.2, "Ice Cream":21.2, 
             "Chocolate Cake":30.3, "Soda":28.9, "Crackers":11.8}


# Data variable
xlabels = food_data.keys()
values = food_data.values()
positions = np.arange(len(values))

# Format y axis
def format_ylabel(valor, posicion):
    return str(int(valor)) + "%" 

# We obtain the Axe object and modify the title and label of the y-axis.
ax = plt.axes()
ax.set_title("Amount of Sugar in Certain Foods")
ax.set_ylabel("Percent of Sugar")

# We format the x-axis labels
# If the word is a compound word, we use two lines
fmt_xlabels = []
for value in xlabels:
    if " " in value:
        fmt_xlabels.append(value.replace(" ", "\n"))
    else:
        fmt_xlabels.append(value)

# We establish major formatters and locators 
ax.yaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_ylabel))
ax.xaxis.set_major_locator(ticker.FixedLocator(positions))
ax.xaxis.set_major_formatter(ticker.FixedFormatter(fmt_xlabels))

#Create the plot
ax.bar(positions, values, color="r", align="center")


#What if we wanted to change the size of the labels?
#One way to do it...
labels = ax.get_xticklabels()
plt.setp(labels, fontsize = 9)

# Or also with
for t in ax.get_xticklabels():
    t.set_fontsize(9)
```

<h4 style="colour: darkgreen">Date formatting with datetime and matplotlib.dates</h4>

The Python **datetime** module contains functionality to handle dates.

>- `datetime(year, month, day[, hour[, minute[, second[, microsecond[, microsecond]]]])`: Creates a datetime object representing a given date (or time).

To use these objects in Matplotlib it is necessary to transform them to decimal numbers. The dates submodule of Matplotlib provides some useful functions to perform these conversions, as well as providing special formatters and locators. (DayFormatter, DayLocator, MonthLocator, HourLocator, etc.)

>- `date2num(datetime)`: Converts an object (or sequence of objects) of type datetime to float values representing time from 0001-01-01 00:00:00 UTC.
>- `num2date(num)`: Converts a number (or sequence) representing time from 0001-01-01 00:00:00 UTC to a datetime object.
>- `drange(dstart, dend, delta)`: This function returns a sequence of numbers representing time between dstart and dend, spaced according to delta. dstart and dend are datetime objects, and delta is a timedelta object.

Another alternative for automatically formatting labels with dates is the function:
- `autofmt_xdate(bottom=0.2, rotation=30, ha=u'right')`: Automatically aligns x-axis date labels.

```python
# Example of formatting dates in a matplotlib chart
# Import datetime and matplotlib.dates modules
import matplotlib.dates as mdates
import datetime as dt

# Create figure and y axis
fig = plt.figure()
ax = fig.add_subplot(111)

# For x we generate daily data between 32/09/2008 and 3/10/2008.
date_1 = dt.datetime(2008, 9, 23)
date_2 = dt.datetime(2008, 10, 3)
delta = dt.timedelta(days=1)
dates = mdates.drange(date_1, date_2, delta)

# For the and random data between 0 and 1
y = np.random.rand(len(dates))

# Plot the data
ax.plot_date(dates, y, linestyle='-')

# We create a formatter object with the day-month-year and apply it to the x-axis.
dateFmt = mdates.DateFormatter('%d-%m-%Y')
ax.xaxis.set_major_formatter(dateFmt)

# For older ticks we will take days
daysLoc = mdates.DayLocator()
ax.xaxis.set_major_locator(daysLoc)

# For minor ticks a period of 6 hours
hoursLoc = mdates.HourLocator(interval=6)
ax.xaxis.set_minor_locator(hoursLoc)

# We automatically adjust the dates of the X-axis.
fig.autofmt_xdate()
```

```python
# Example with month data

# Figure and one axis
fig = plt.figure()
ax = fig.add_subplot(111)

# Temporary data every 10 days from 23/9/2008 to 16/2/2009
date_1 = dt.datetime(2008, 9, 23)
date_2 = dt.datetime(2009, 2, 16)
delta = dt.timedelta(days=10)
dates = mdates.drange(date_1, date_2, delta)

# Random y values
y = np.random.rand(len(dates))

ax.plot_date(dates, y, linestyle='-')

# Locators months and week for x
monthsLoc = mdates.MonthLocator()
weeksLoc = mdates.WeekdayLocator()

# Apply to x
ax.xaxis.set_major_locator(monthsLoc)
ax.xaxis.set_minor_locator(weeksLoc)

# A simple formatter for data
monthsFmt = mdates.DateFormatter('%b')
ax.xaxis.set_major_formatter(monthsFmt)
```

<h3 style="colour: navy">Controlling a grid</h3>.

To display a grid we will use the grid() method.
>- `grid(b=None, which='major', axis='both', **kwargs)`
> - **_b_**: Specifies whether to display the grid (True) or not (False). For compatibility with matlab it also accepts "on" and "off".
> - **_wich_**: Specifies whether to show "major", "minor" or "both"</li>.
> - **_axis_**: Specifies whether to display "x", "y" or "both".
> - **_kwargs_**: Style arguments for grid lines

```python
# Plots normalised and denormalised cardinal sine functions between -20 and 20
# Creates a grid greater than 5 for the x-axis
# Creates a grid greater than 0.2 for the y-axis

X = np.linspace(-20, 20, 1024)
Y1 = np.sinc(X)
Y2 = np.sin(X)/X
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.2))
plt.grid(True)
plt.plot(X, Y1, c = 'r')
plt.plot(X, Y2, c = "b")
```

```python
# Represents the above function
# Represents grids greater than 5 and 0.2 for the x and y axes with a grey line at 70%.
# Represents a grid less than 1 for the x-axis with a dashed grey line at 40%. 
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.2))
plt.grid(True, which='minor', axis="x", linestyle ="--", c="0.4", lw=0.25)
plt.grid(True, which='major', axis="both", linestyle ="-", c="0.7", lw=0.5)

plt.plot(X, Y1, c = 'r')
plt.plot(X, Y2, c = "g")
```

```python
# Add legend and title to the above graphic
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.2))
plt.grid(True, which='minor', axis="x", linestyle ="--", c="0.4", lw=0.25)
plt.grid(True, which='major', axis="both", linestyle ="-", c="0.7", lw=0.5)

plt.title("Funciones seno cardinal")
plt.xlabel("X")
plt.ylabel("Y")

label1 = "$\\frac{sin(\pi x)}{ \pi x}$"
label2 = "$\\frac{sin(x)}{ x}$"

plt.plot(X, Y1, c='r', label=label1)
plt.plot(X, Y2, c="g", label=label2)
plt.legend(fontsize="14")
```

<h3 style="colour: navy">Working with colour ramps</h3>.

To work with colour ramps, we can use the cm module of matplotlib. Some of the matplotlib functions allow you to specify a colour ramp via the cmap parameter. We can also get colours from a colour ramp, once we have created it.
>- `cm.get_cmap(name)`.
>- Create a ColorMap object. To retrieve colours we simply call:
> - `cmap(n)` : Where n is a number ranging from 0 (first colour of the ramp) to 1 (last colour of the ramp).

```python
# Create 20 lines according to the equations y = x + n (where n [0,2,4,6, ...])
# Give each line a different colour according to the colour ramp "plasma".

import matplotlib.cm as cm
#Creamos una rampa de color
cmap = cm.get_cmap("plasma")
x = np.arange(0,26)
fig, ax = plt.subplots()
for n in range(20):
    cind = n/20.
    color = cmap(cind)
    ax.plot(x, x+2*n, c = color)
```
