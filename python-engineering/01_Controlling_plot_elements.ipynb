<h2 style="color: olive">Controlando los elementos de nuestro gráfico</h2>

Matplotlib ofrece una gran libertad para configurar opciones del propio gráfico, como títulos, leyenda, limites de ejes, etc. Una gran fortaleza de matplotlib es que incluye un motor para renderizar códigos de Latex, los cuales podemos incluir en cualquier parte de nuestro gráfico

```python
# Importamos los módulos necesarios
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

***
<h3 style="color: navy">Títulos de gráfico y etiquetas de los ejes</h3>

Las siguientes funciones permiten especificar títulos para el gráfico y sus ejes x e y:

>- `title(title)` Título del grafico. Acepta una cadena de texo.
>- `xlabel(label)` Título del eje x. Acepta una cadena de texo.
>- `ylabel(label)` Título del eje y. Acepta una cadena de texo.
>
> Las cadenas de texto pueden incorporar códigod de latex (los encerraremos entre símbolos de 'dolar' '$'. Si nuestra cadena de texto lleva tildes o caracteres especiales, ponderemos una u delante (unicode)

```python
# Carga los datos de "data/iris.dat"
# Estos datos provienen del dataset de Fisher disponible en 
# http://archive.ics.uci.edu/ml/datasets/Iris
# El dataset tiene siguientes columnas separadas por comas
# 1. sepal length in cm 
# 2. sepal width in cm 
# 3. petal length in cm 
# 4. petal width in cm 
# 5. class: Iris Setosa - Iris Versicolour - Iris Virginica
# Representa un gráfico con las columnas 1 y 2 con colores diferentes para cada clase
# Añade títulos de gráfico y etiquetas para los ejes

labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colores = ["r", "b", "g"]
iris_data = np.genfromtxt("data/iris.dat", delimiter=",", dtype=None)
clases = iris_data["f4"]

for idx, label in enumerate(labels):
    s_lengths = iris_data["f0"]
    s_widths = iris_data["f1"]
    ids = clases == label
    plt.scatter(s_lengths[ids], s_widths[ids], marker="o", color=colores[idx])

plt.title("Fisher's Iris Dataset")
plt.xlabel("Sepal length (cm)")
plt.ylabel("Petal length (cm)")
```

```python
# Crea una gráfica del polinómio f(x) = 1/4 * (x+4)(x+1)(x-2) 
# entre los valores de x [-4, 4]
# Represéntalo con una línea verde
# Añade título a gráfico y etiquetas a ejes
x = np.linspace(-4, 4, 25)
y = (1.0/4) * (x+4)*(x+1)*(x-2)
plt.plot(x,y, color = "g")
plt.title('$f(x)=\\frac{1}{4}(x+4)(x+1)(x-2)$')
plt.xlabel("X")
plt.ylabel("Y")
```

<h3 style="color: navy">Leyenda</h3>

Para añadir etiquetas a las líneas que representemos con plot, especificaremos el argumento **label** cuando llamemos a plot. Cuando llamemos a la función **legend** está mostrará el estilo de línea y el label especificado.

Para mostrar la leyenda, utilizaremos a la función **plt.legend(args, kwargs)**. Si hemos especificado el argumento *label* estos se mostrarán en la leyenda, sino es posible especificar que se mostrará utilizando un array con de líneas y otro de etiquetas.

> `plt.legend(*args, **kwargs)`
>- Los argumentos (si los especificamos) serán una lista de líneas y otra de etiquetas de la siguiente forma: `plt.legend([line1, line2, ..., linen], [label1, label2, ..., labeln]`
>- Alguno de los keyword arguments que podemos utilizar son:
>    - **loc**: Entero que especifica la localización de la leyenda. Los códigos posibles son; ['best':0, 'upper right':1, 'upper left': 2, 'lower left':3, 'lower right':4, 'right':5, 'center left':6, 'center right':7, 'lower center':8, 'upper center':9, 'center':10]

>    - **bbox_to_anchor**: Tupla con las coordenadas (0-1) de la localización de la leyenda. Por ejemplo, para colocar la leyenda en la parte central del eje de la derecha utilizaremos:
>        - `loc='upper right', bbox_to_anchor=(0.5, 0.5)`
>    - **ncol** : Entenro que especifica el número de columnas de la leyenda (1 por defecto)
    The number of columns that the legend has. Default is 1.
>    - **fontsize** : Tamaño de la funente de la leyenda
>    - **frameon** : None o (True/False). Especifica si se dibuja una caja alrededor de la leyenda 
>    - **fancybox** : None o (True/False). Especifica si la caja de la leyenda tiene los bordes redondeados. (solo si frameon == True)
>    - **shadow** : None o (True/False). Especifica si la caja de la leyenda tendrá sombra.
>    - **framealpha** : Float que especifica la transparencia de la caja de la leyenda
>    - **title** : Cadena de texto con el título de la leyenda.

```python
# Crea una representación de las siguiente funciones entre [0, 100]
# y1 = x^2 -4x +10  --> línea roja discontinua y puntos azules
# y2 = -x^2 +4x + 10 --> línea azul discontinua y puntos rojos
# Añade un título al gráfico y una leyenda

x = np.linspace(0, 100, 20)
y1 = x**2 - 4*x + 10
y2 = -x**2 + 4*x + 10
func_label_1 = "$f(x)=x^2 -4x + 10$"
func_label_2 = "$f(x)=-x^2 +4x + 10$"
plt.plot(x, y1, c="r", mfc="b", marker="o", mec="k", label=func_label_1)
plt.plot(x, y2, c="b", mfc="r", marker="o", mec="k", label=func_label_2)
plt.title(u"Representación de dos polinómios")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(title=u"Polinómios", shadow=True, loc=3, fancybox=True)
```

```python
# En este ejemplo vamos a llamar a la leyenda de otra forma
# Guardaremos las líneas en variables y las utilizaremos al llamar a la leyenda

line1, = plt.plot(x, y1, c="r", mfc="b", marker="o", mec="k")
line2, = plt.plot(x, y2, c="b", mfc="r", marker="o", mec="k")
plt.title(u"Representación de dos polinómios")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend((line1, line2), (func_label_1, func_label_2), title=u"Polinómios", 
           shadow=True, loc=3, fancybox=True)
```

```python
# Repite la gráfica anterior pero saca la leyenda fuera de área del gráfico
# Utiliza la propiedad bbox_to_anchor

line1, = plt.plot(x, y1, c="r", mfc="b", marker="o", mec="k")
line2, = plt.plot(x, y2, c="b", mfc="r", marker="o", mec="k")
plt.title(u"Representación de dos polinómios")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend((line1, line2), (func_label_1, func_label_2), title=u"Polinómios", 
           shadow=True, loc='lower left', bbox_to_anchor=(1, 0))
```

#### Control de ticks con el módulo ticker

El módulo ticker contiene clases para controlar la localización y el formato de los ticks. Para cambiar la localización y el formato de los ticks podemos utilizar las funciones del objeto Axe:

- `axe.xaxis.set_major_locator(locator)`: Acepta un objeto *matplotlib.ticker.locator* para los ticks principales.
- `axe.xaxis.set_minor_locator(locator)`: Acepta un objeto "matplotlib.ticker.locator" para los ticks secundarios
- `axe.xaxis.set_major_formatter(formatter)`: Acepta un objeto "fomatter" para los ticks principales.
- `axe.xaxis.set_minor_formatter(formatter)`: Acepta un objeto "formatter" para los ticks secundarios.

Para crear objetos locator, podemos utilizar las funciones del módulo ticker:
- `ticker.MultipleLocator(d)`:  Crea ticks en el intervalo especificado por d
- `ticker.FixedLocator(positions)`: Crea ticks en las posiciones especificadas en la lista positions
- `ticker.FixedFormatter(namelist)`: Crea etiquetas de acuerdo a una lista de nombres
- `ticker.FuncFormatter(function)`: Crea etiquetas de acuerdo a una función

```python
# Representa las funciones de seno cardinal normalizada y desnormalizada entre -20 y 20
# https://en.wikipedia.org/wiki/Sinc_function
# Numpy proporciona la función normalizada np.sinc()
# Para la funcion desnormalizada utilizaremos sin(X)/X
# Usa en el eje X ticks mayores de 5 y menores de 1
# Usa en el eje Y ticks mayores de 0.2 y menores de 0.1
import matplotlib.ticker as ticker
X = np.linspace(-20, 20, 1024)
Y1 = np.sinc(X)
Y2 = np.sin(X)/X

# Para utilizar estas las funciones del módulo ticker debemos recuperar el objeto Axe
ax = plt.axes()
# Cambiamos los ticks de X
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.1))
plt.plot(X, Y1, c = 'r')
plt.plot(X, Y2, c = "b")
```

```python
# Crea un gráfico de barras simple con los valores y etiquetas de las siguientes listas
values = (25, 50, 78, 30, 28)
labels = ('Vicente', 'Maria', 'Alejandro', 'Alberto', 'Jose')

pos_list = np.arange(len(labels))
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.FixedLocator((pos_list)))
ax.xaxis.set_major_formatter(ticker.FixedFormatter((labels)))

# Creamos el gráfico de barras (con el argumento align='center')
plt.bar(pos_list, values, color='r', align='center')
```

```python
# Representa las funciones y = e^-10x e y = e^-5x entre 0 y 1
# Utiliza una linea negra y otra negra discontinua
# Usa ticker.FuncFormatter() para formatear las etiquetas del eje x segun make_label()
def make_label(value, pos):
    return str(value * 100) + "%"

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(make_label))
X = np.linspace(0, 1, 100)
plt.plot(X, np.exp(-10 * X), c ='k')
plt.plot(X, np.exp(-5 * X), c= 'k', ls = '--')
```

```python
# Rota las etiquetas anteriores 30 grados
# Utiliza el método get_xtickslabels() del objeto Axe para recuperar las etiquetas
# Utiliza el método set_rotation(grados) de cada objeto etiqueta para rotarlas
def make_label(value, pos):
    return str(value * 100) + "%"

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(make_label))
X = np.linspace(0, 1, 100)
plt.plot(X, np.exp(-10 * X), c ='k')
plt.plot(X, np.exp(-5 * X), c= 'k', ls = '--')

#Vamos a rotar las etiquetas de x 30 grados
labels = ax.get_xticklabels()
for label in labels:
    label.set_rotation(30.)
```

```python
# El siguiente diccionario contiene las cantidades de azucar (%) de varios alimentos
# Realiza un gráfico de barras con título y etiquetas de los ejes
# Las etiquetas del eje y deberán incorporar "%"
food_data = {"Ketchup":8.8, "Penaut Butter":9.2, 
             "Chocolate Bar":33.2, "Ice Cream":21.2, 
             "Chocolate Cake":30.3, "Soda":28.9, "Crackers":11.8}


# Variables para los datos
xlabels = food_data.keys()
values = food_data.values()
positions = np.arange(len(values))

# Función para formatear las etiquetas del eje y
def format_ylabel(valor, posicion):
    return str(int(valor)) + "%" 

# Obtenemos el objeto Axe y modificamos el título y la etiqueta del eje y
ax = plt.axes()
ax.set_title("Amount of Sugar in Certain Foods")
ax.set_ylabel("Percent of Sugar")

# Formateamos las etiquetas del eje x
# Si la palabra es compuesta, utilizaremos dos líneas
fmt_xlabels = []
for value in xlabels:
    if " " in value:
        fmt_xlabels.append(value.replace(" ", "\n"))
    else:
        fmt_xlabels.append(value)

# Establecemos major formatters y locators 
ax.yaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_ylabel))
ax.xaxis.set_major_locator(ticker.FixedLocator(positions))
ax.xaxis.set_major_formatter(ticker.FixedFormatter(fmt_xlabels))

#Creamos el gráfico
ax.bar(positions, values, color="r", align="center")

#Y si quisieramos cambiar el tamaño de las etiquetas
#Una forma de hacerlo...
labels = ax.get_xticklabels()
plt.setp(labels, fontsize = 9)

#Otra forma
for t in ax.get_xticklabels():
    t.set_fontsize(9)
```

<h4 style="color: darkgreen">Formato de fechas con datetime y matplotlib.dates</h4>

El módulo **datetime** de Python contiene funcionalidad para manejar fechas.

>- `datetime(year, month, day[, hour[, minute[, second[, microsecond]]]])`: Crea un objeto datetime que representa una fecha (o momento) determinado

Para utilizar estos objetos objetos en Matplotlib es necesario transformarlos a números decimales. El submódulo dates de Matplotlib proporciona algunas funciones útiles para realizar estas conversiones, así como también proporciona formatters y locators especiales. (DayFormatter, DayLocator, MonthLocator, HourLocator, etc.)

>- `date2num(datetime)`: Convierte un objeto (o secuencia de objetos) de tipo datetime a valores flotantes que representan tiempo desde el 0001-01-01 00:00:00 UTC.
>- `num2date(num)`: Convierte un numero (o secuencia) que representan tiempo desde el 0001-01-01 00:00:00 UTC a un objeto datetime.
>- `drange(dstart, dend, delta)`: Esta función devuelve una secuencia de números que representan tiempo entre dstart y dend, espaciados según delta. dstart y dend son objetos datetime, y delta es un objeto de tipo timedelta.

Otra alternativa para formatear automaticamente etiquetas con fechas es la función:
- `autofmt_xdate(bottom=0.2, rotation=30, ha=u'right')`: Alinea automaticamente etiquetas de fechas del eje x

```python
# Ejemplo de formateado de fechas en un gráfico de matplotlib
# Importamos los módulos datetime y matplotlib.dates
import matplotlib.dates as mdates
import datetime as dt

# Creamos una figura y un axe
fig = plt.figure()
ax = fig.add_subplot(111)

# Para la x generamos datos diarios entre el 32/09/2008 y el 3/10/2008
date_1 = dt.datetime(2008, 9, 23)
date_2 = dt.datetime(2008, 10, 3)
delta = dt.timedelta(days=1)
dates = mdates.drange(date_1, date_2, delta)

# Para la y datos aleatorios entre 0 y 1
y = np.random.rand(len(dates))

# Ploteamos los datos
ax.plot_date(dates, y, linestyle='-')

# Creamos un objeto formatter con dia-mes-año y lo aplicamos al eje x
dateFmt = mdates.DateFormatter('%d-%m-%Y')
ax.xaxis.set_major_formatter(dateFmt)

# Para los tick mayores cogeremos días
daysLoc = mdates.DayLocator()
ax.xaxis.set_major_locator(daysLoc)

# Para los tick menores un periodo de 6 horas
hoursLoc = mdates.HourLocator(interval=6)
ax.xaxis.set_minor_locator(hoursLoc)

# Ajustamos automaticamente las fechas del eje X
fig.autofmt_xdate()
```

```python
# Ejemplo con datos mensuales

# Creamos una figura con un Axe
fig = plt.figure()
ax = fig.add_subplot(111)

# Datos temporales cada 10 días desde el 23/9/2008 al 16/2/2009
date_1 = dt.datetime(2008, 9, 23)
date_2 = dt.datetime(2009, 2, 16)
delta = dt.timedelta(days=10)
dates = mdates.drange(date_1, date_2, delta)

# Valores de y aleatorios
y = np.random.rand(len(dates))

# Ploteamos los datos
ax.plot_date(dates, y, linestyle='-')

# Locators mensuales y semanales para el eje X
monthsLoc = mdates.MonthLocator()
weeksLoc = mdates.WeekdayLocator()

# Los aplicamos al eje X
ax.xaxis.set_major_locator(monthsLoc)
ax.xaxis.set_minor_locator(weeksLoc)

# Un formatter simple para los datos
monthsFmt = mdates.DateFormatter('%b')
ax.xaxis.set_major_formatter(monthsFmt)
```

<h3 style="color: navy">Control de un grid</h3>

Para mostrar un grid utilzaremos el método grid()
>- `grid(b=None, which='major', axis='both', **kwargs)`
>    - **_b_**: Especifica si mostrar el grid (True) o no (False). Por compatibilidad con matlab tambien acepta "on" y "off".
>    - **_wich_**: Especifica si mostrar "major", "minor" o "both"</li>
>    - **_axis_**: Especifica si mostrar "x", "y" o "both"
>    - **_kwargs_**: Argumentos de estilos para las líneas del grid

```python
# Representa las funciones de seno cardinal normalizada y desnormalizada entre -20 y 20
# Crea un grid mayor de 5 para el eje x
# Crea un grid mayor de 0.2 para el eje y

X = np.linspace(-20, 20, 1024)
Y1 = np.sinc(X)
Y2 = np.sin(X)/X
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.2))
plt.grid(True)
plt.plot(X, Y1, c = 'r')
plt.plot(X, Y2, c = "b")
```

```python
# Representa la función anterior
# Representa grid mayores de 5 y 0.2 para los ejes x e y con una línea gris al 70%
# Representa un grid menor de 1 para el eje x con una línea discontinua gris al 40% 
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.2))
plt.grid(True, which='minor', axis="x", linestyle ="--", c="0.4", lw=0.25)
plt.grid(True, which='major', axis="both", linestyle ="-", c="0.7", lw=0.5)

plt.plot(X, Y1, c = 'r')
plt.plot(X, Y2, c = "g")
```

```python
# Añade leyenda y título al gráfico anterior
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.2))
plt.grid(True, which='minor', axis="x", linestyle ="--", c="0.4", lw=0.25)
plt.grid(True, which='major', axis="both", linestyle ="-", c="0.7", lw=0.5)

plt.title("Funciones seno cardinal")
plt.xlabel("X")
plt.ylabel("Y")

label1 = "$\\frac{sin(\pi x)}{ \pi x}$"
label2 = "$\\frac{sin(x)}{ x}$"

plt.plot(X, Y1, c='r', label=label1)
plt.plot(X, Y2, c="g", label=label2)
plt.legend(fontsize="14")
```

<h3 style="color: navy">Añadir anotaciones</h3>

Para crear anotaciones en la gráfica con flechas utilizaremos
>- `annotate(s, xy, xytext, arrowprops)`
>    - s :Cadena de texto de la anotación
>    - xy : Posición del elemento a anotar (extremo final de flecha, fig. coord.)
>    - xytext : Posición del texto (etiqueta) (coordenadas de figura)
>    - arrowprops : Diccionario con las propiedades de la flecha

> Algunos de los elementos que podemos especificar en **arrowprops** son:
>    - widt : Ancho de la flecha
>    - frac : Fracción de la flecha ocupada por la cabeza
>    - shrink : Separa la flecha del punto  (defecto 0.05, un 5%)

<h3 style="color: navy">Trabajo con rampas de color</h3>

Para trabajar con rampas de color, podemos utilizar el módulo cm de matplotlib. Algunas de las funciones de matplotlib permiten especificar una rampa de color a través del parámetro cmap. También podemos obtener colores de una rampa de color, una vez la hemos creado.
>- `cm.get_cmap(name)`
>- Crea un objeto ColorMap. Para recuperar colores simplemente llamaremos:
>    - `cmap(n)` : Donde n es un número que va de 0 (primer color de la rampa) a 1 (último color de la rampa)

```python
# Crea 20 líneas según las ecuaciones y = x + n (siendo n [0,2,4,6, ...])
# Dale a cada línea un color diferente según la rampa de color "plasma"

import matplotlib.cm as cm
#Creamos una rampa de color
cmap = cm.get_cmap("plasma")
x = np.arange(0,26)
fig, ax = plt.subplots()
for n in range(20):
    cind = n/20.
    color = cmap(cind)
    ax.plot(x, x+2*n, c = color)
```
